```python
def get_nutrient_df(all_dict, df_name="References/2007-Bang thanh phan dinh duong VN.pdf"):
    # read pdf file
    doc = fitz.open(df_name)
    all_rows = []
    for page in doc[13:]:
        tabs = page.find_tables()
        # skip if there is no table
        try:
            tab = tabs[0]
        except IndexError:
            continue
        df = tab.to_pandas()
        # get row as a list of value
        row = df.iloc[:, 2][0].split('\n')
        row.pop(2)
        all_rows.append(row[1:5])
    # create dataframe
    nutrient_df = pd.DataFrame(all_rows, columns=['Energy', 'Protein', 'Fat', 'Carbohydrate'])
    nutrient_df.index += 1

    name_lookup = {}
    group_lookup = {}
    # add name and group columns
    for i, d in enumerate(all_dict):
        for k, v in d.items():
            name_lookup[k] = v
            group_lookup[k] = f"Group_{i+1}"
    nutrient_df["Name"] = nutrient_df.index.map(name_lookup)
    nutrient_df["Group"] = nutrient_df.index.map(group_lookup)

    return nutrient_df
```

```python
df = get_nutrient_df(all_dict)
```
